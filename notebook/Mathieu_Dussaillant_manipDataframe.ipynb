```python
# préciser le chemin du fichier à lire
cheminDuFichierCSV = "./../data/dictionary/Buchanan_PronChar_counts.csv" 

# demander à pandas de lire le fichier et garder ce tableau accessible dans une variable
dfDicoPronChar = pd.read_csv(cheminDuFichierCSV, delimiter=";", encoding="utf8")

# afficher à l'écran un échantillon du tableau
dfDicoPronChar.sample(n=5)
```

```python
dfDicoPronChar = dfDicoPronChar.dropna()
```

```python
def sampa2ipa(strSampa, dfMapping):
    ipaRes = ""
    
    #print(listValSampa)
    for valSampa in strSampa:
        dfRowFinded = dfMapping[dfMapping['char']==valSampa]

        if len(dfRowFinded==1):
            ipaRes+=str(dfRowFinded.iloc[0]['graph'])
        else:
            ipaRes+=''
    return ipaRes
```

```python
dfDicoBigiRich = dfDicoBuch.sample(n=100).copy()

dfDicoBuch['headword'] = dfDicoBuch.apply(lambda x: sampa2ipa(x['pron'], dfDicoPronChar),1)
```
